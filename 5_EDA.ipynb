```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student-mat.csv", sep=";")

print(df.describe())
print(df.isnull().sum())

plt.hist(df["G3"],bins=20)
plt.title("Distribution of Final Grade (G3)")
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.show()

plt.scatter(df["G1"],df["G3"])
plt.xlabel("G1")
plt.ylabel("G3")
plt.title("G1 vs Final Grade")
plt.show()
```
